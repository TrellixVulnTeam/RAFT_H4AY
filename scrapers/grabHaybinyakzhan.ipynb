```python
import requests
from pathlib import Path
import os
from bs4 import BeautifulSoup
```

```python
links = {}
for year in range(2010,2020):
    for month in range(1,13):
        y = str(year)
        m = str(month)
        if (month < 10):
            m = '0'+m
        r = requests.get('http://haybinyakzhan.blogspot.com/'+str(y)+'/'+str(m)+'/')
        soup = BeautifulSoup(r.text, 'html.parser')
        menu = soup.find('div',{'id':'ArchiveList'})
        uls = menu.find_all('ul',{'class':'posts'})
        for ul in uls:
            aTags = ul.find_all('a')
            for a in aTags:
                href = a.get('href')
                links[href] = False
print(len(links))
```

```python
directory = 'C:\\Users\\maste\\Desktop\\freqCorpus\\haybinyakzhan'
comments = ''
count = 0
for link in links:
    print(link)
    r = requests.get(link)
    soup = BeautifulSoup(r.text, 'html.parser')
    body = soup.find('div', {'class':'post-body'})
    textArr = body.text.split('\n')
    text = ''
    for t in textArr:
        if (t != ''):
            text += (t + '\n')
    if (text != ''):
        file = os.path.join(directory, 'haybinyakzkhan_'+str(count)+'.txt')
        print(file)
        with open(file, "w+", encoding="utf8") as f:
            f.write('<difficulty></difficulty>\n')
            f.write('<body>\n')
            f.write(text)
            f.write('\n</body>')
        count += 1
    commentDiv = soup.find('div', {'class':'comments'})
    commentText = commentDiv.text.split('\n')
    for t in commentText:
        if (t != ''):
            comments += t
```
