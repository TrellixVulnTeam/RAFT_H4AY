```python
import requests
from pathlib import Path
import os
from bs4 import BeautifulSoup
```

```python
urlList = {}
r = requests.get('https://mawdoo3.com/')
soup = BeautifulSoup(r.text, 'html.parser')
uls = soup.find_all('ul', {'class':'category-items'})
if (uls != None):
    for ul in uls:
        links = soup.find_all('a')
        for link in links:
            href = link.get('href')
            if (href != None):
                if (href.find('http') == -1 and href != '/'):
                    url = 'https://mawdoo3.com/' + href
                    if (url.find(':') == -1):
                        urlList[url] = False
                    else:
                        r = requests.get(url)
                        soup = BeautifulSoup(r.text, 'html.parser')
                        ul = soup.find('ul', {'class':'categories-list'})
                        if (ul != None):
                            lis = ul.find_all('li')
                            for li in lis:
                                a = li.find('a',{'class':'category-box'})
                                if (a != None):
                                    href = a.get('href')
                                    if (href != None):
                                        url = 'https://mawdoo3.com' + href
                                        urlList[url] = False
                    print(len(urlList))
                        
```

```python
directory = 'C:\\Users\\maste\\Desktop\\freqCorpus\\mawdoo'
count = 0
for url in urlList:
    done = urlList[url]
    if (done == False):
        count += 1
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')
        article = soup.find('div',{'id':'article-box'})
        if (article != None):
            for script in article(["script", "style"]):
                script.decompose()    # rip it out
            articleText = article.text
            text = ''
            lines = articleText.split('/n')
            for line in lines:
                if (line != ''):
                    text += (line + '\n')
            name = soup.find('h1',{'class':'title'}).text
            name = name.replace(' ','_')
            file = os.path.join(directory, name+'.txt')
            with open(file, "w+", encoding="utf8") as f:
                f.write('<difficulty></difficulty>\n')
                f.write('<body>\n')
                f.write(text)
                f.write('\n</body>')
            urlList[url] = True
```
