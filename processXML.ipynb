```python
from bs4 import BeautifulSoup
import lxml
from pathlib import Path
import os
```

```python
def standardizeWord(word):
    newWord = ''
    for i in range(0,len(word)):
        char = word[i]
        newWord += standardizeChar(char)
    return newWord

def standardizeChar(char):
    s = {}
    s['أ']='ا'
    s['ٱ']='ا'
    s['آ']='ا'
    s['ؤ']='و'
    if char in s:
        return s[char]
    else:
        return char
```

```python
directory = 'C:\\Users\\maste\\Desktop\\RAFT\\arabicScraperOutput\\xml\\tagged'
```

```python
for filename in os.listdir(directory):
    if filename.endswith('.xml'):
        print(filename)
        myFile = os.path.join(directory, filename)
        f = open(myFile, "r", encoding="utf8")
        text = f.read()
        f.close()
        
        soup = BeautifulSoup(text,'html.parser')
        words = soup.find_all('word')
        if (len(words) > 0):
            myFileOut = os.path.join('C:\\Users\\maste\\Desktop\\RAFT\\arabicScraperOutput\\frequency_list',filename[:-4]+'.txt')
            fout = open(myFileOut, "w+",encoding='utf8')
            for w in words:
                analysis = w.find('analysis')
                if analysis != None:
                    data = analysis.find('morph_feature_set')
                    pos = data['pos']
                    if pos != 'punc':
                        word = w['word']
                        gloss = data['gloss']
                        lemmaFull = data['lemma']
                        lemma = ''
                        for l in lemmaFull:
                            #1568 - 1610 discludes vowels
                            #1536 - 1791 includes vowels
                            if (ord(l) >= 1648 and ord(l) <= 1651):
                                l = 'ا'
                            if (ord(l) >= 1568 and ord(l) <= 1610): 
                                lemma += l
                        fout.write(word+":::"+lemma+":::"+gloss+":::"+pos+'\n')
            fout.close()
print('finished')
```
