```python
def switchToFrame(driver, name):
    driver.switch_to.default_content()
    driver.switch_to.frame(driver.find_element_by_name(name))
```

```python
driver = webdriver.Chrome()
driver.get("https://corpus.byu.edu/iweb/")
driver.implicitly_wait(1)
```
