```python
from bs4 import BeautifulSoup
import lxml
from pathlib import Path
import os
import pandas as pd
import re
import sys
from doc import Doc, standardizeWord, standardizeChar
from io import StringIO
```

```python
directory = 'C:\\Users\\maste\\Desktop\\RAFT\\freqCorpus\\tagged'
last = ''
skip = False
freqList = {};
count = 0
```

```python
for filename in os.listdir(directory):
    if skip == False:
        if filename.endswith('.xml'):
            skip = True
            file = os.path.join(directory, filename)
            with open(file, "r", encoding="utf8") as f:
                doc = f.read()
                soup = BeautifulSoup(doc, 'html.parser')
                words = soup.find_all('analysis')
                for word in words:
                    data = word.find('morph_feature_set')
                    lemma = data.get('lemma')
                    pos = data.get('pos')
                    if (pos != 'punc'):
                        key = lemma +":::"+pos
                        if (key in freqList):
                            freqList[key] += 1
                        else:
                            freqList[key] = 1
            count += 1
            last = filename
            print(filename, ":", len(freqList), ",", count)
            skip = False
    else:
        print("skipped ", filename)
        if filename == last:
            skip = False
```
