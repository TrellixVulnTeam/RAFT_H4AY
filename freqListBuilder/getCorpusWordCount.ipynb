```python
from pathlib import Path
import os
from bs4 import BeautifulSoup
```

```python
def wordCount(doc):
    docArray = doc.split()
    count = 0
    for i in docArray:
        if i != '':
            count += 1
    return count

def getWordCount(file):
    count = 0
    with open(file, "r", encoding="utf8") as f:
            doc = f.read()
            soup = BeautifulSoup(doc, 'html.parser')
            body = soup.find('body')
            if (body != None):
                t = body.text
            else:
                t = doc
            c = wordCount(t)
            count += c
    return count

def getFiles(directory):
    count = 0
    for fname in os.listdir(directory):
        file = os.path.join(directory, fname)
        if (os.path.isdir(file) == True):
            count += getFiles(file)
        else:
            count += getWordCount(file)
    print(directory, count)
    return count
```
