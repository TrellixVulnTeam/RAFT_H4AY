```python
def convertChar(charIn):
    return conversion[charIn]

def convertString(stringIn):
    newString = ""
    for char in stringIn:
        if char in conversion:
            newString += convertChar(char)
    return newString

def standardizeWord(word):
    newWord = ''
    for i in range(0,len(word)):
        char = word[i]
        newWord += standardizeChar(char)
    return newWord

def standardizeChar(char):
    s = {}
    s['أ']='ا'
    s['ٱ']='ا'
    s['آ']='ا'
    s['ؤ']='و'
    if char in s:
        return s[char]
    else:
        return char
```
